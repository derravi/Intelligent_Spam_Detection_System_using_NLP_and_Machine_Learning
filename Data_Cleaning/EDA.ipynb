```python
import pandas as pd
import numpy as np
```

```python
try:

    df = pd.read_csv("../Data_set/spam.csv", encoding='latin-1')
    print("File loaded successfully")
    
except FileNotFoundError as e:
    print(f"File not found: {e}.")
except Exception as e1:
    print(f"Something went wrong: {e1}.")
    
```

```python
df.head()
```

```python
print(f"Les see the Old data Into this CSV file they have total {df.shape[0]} rows and the {df.shape[1]} columns.")
```

```python
#Lets Remove the Unwanted Columns from the CSV File.

print("Lets remove the unwanted columns.")
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)

#Recheck the total rows and columns
print(f"Lets see updated csv file they have a total {df.shape[0]} rows and the {df.shape[1]} columns.")

#lets change the columns sequance and the name of the columns.

print("lets change the columns sequance and the name of the columns.")
df = df.rename(columns={'v1':'output','v2':'input'})

#Lets change the sequance of the columns.
df.columns = ['output','input']
df = df[['input','output']]
```

```python
#lets see the null values
print("lets see the null values.")
df.isna().sum()
```

```python
print(f"Null values of the input columns is: {df['input'].isna().sum()}")
print(f"Null values of the output columns is: {df['output'].isna().sum()}")
```

```python
unwanted_charecters = r'[.><+\-*/=!@#$%^&()\-:;?`~]'

df['input'] = df['input'].str.lower()
df['input'] = df['input'].replace(unwanted_charecters," ",regex=True)
df['input'] = df['input'].str.strip()
df['input'] = df['input'].replace(r'\s+',' ',regex=True)
df['input'] = df['input'].replace(r'\d+',' ',regex=True)
```

```python
df['output'] = df['output'].replace("ham","human")
```

```python
df.to_csv("../Data_set/updated.csv")
```
